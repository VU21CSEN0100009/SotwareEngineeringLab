Implement weather modeling using the quadratic solution with hard coding variables (fixed variable values give in program)

```python
import matplotlib.pyplot as plt
import numpy as np
a = 2
b = 4
c = -6
x = np.linspace(-5, 5, 200)
y = a*x**2 + b*x + c
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'Hardcoded', color='red')
plt.title('Plot of another quadratic function')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.show()
```

Implement weather modeling using the quadratic solution with hard coding variables through keyboard input by user

```python
import matplotlib.pyplot as plt
import numpy as np
a, b, c = map(int, input("Enter the values of a, b, and c separated by spaces: ").split())
x = np.arange(-10, 10, 0.1)
y = a*x**2 + b*x + c
plt.figure(figsize=(10, 8))
plt.scatter(x, y, label=f'UserInput', color='green')
plt.title('Scatter plot of the quadratic function')
plt.xlabel('Time')
plt.ylabel('Temperature')

plt.legend()
plt.grid(True)
plt.show()
```

Implement weather modeling using the quadratic solution with hard coding variables read from a file


```python
import matplotlib.pyplot as plt
import numpy as np

# Open the file and read the coefficients
with open('Input.txt', 'r') as file:
    d, e, f = map(float, file.readline().split())

# Define the range of x
x = np.linspace(-20, 20, 800)

# Define the quadratic function
y = d*x**2 + e*x + f

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'FileInput')
plt.title('Plot of the modified quadratic function')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.show()
```

Implement weather modeling using the quadratic solution with hard coding variables for multiple sets of inputs
(multiple sets of variable values given in program or file or user input and print all sets of graphs in one plot and specify those sets values)


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the file name
files = 'Input2.txt'

# Define the range of x
x = np.linspace(-20, 20, 800)

# Create the plot
plt.figure(figsize=(8, 6))

# Define the color scheme
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

with open(files, 'r') as file:
  for i, line in enumerate(file):
    a, b, c = map(float, line.split())
    y = a*x**2 + b*x + c
    plt.plot(x, y, label=f'FileInput Line {i+1}: {a}x^2 + {b}x + {c}', color=colors[i % len(colors)])

plt.title('Plot of the modified quadratic functions')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.show()
```

Implement weather modeling using the quadratic solution with combination of hard coding variables (fixed) and user input through keyboard and specify the type of graphs with values



```python
import matplotlib.pyplot as plt
import numpy as np

# First code
a = 2
b = 4
c = -6
x1 = np.linspace(-5, 5, 200)
y1 = a*x1**2 + b*x1 + c

# Second code
a, b, c = map(int, input("Enter the values of a, b, and c separated by spaces: ").split())
x2 = np.arange(-10, 10, 0.1)
y2 = a*x2**2 + b*x2 + c

# Plot both curves on the same figure
plt.figure(figsize=(10, 8))
plt.plot(x1, y1, label=f'Hardcoded', color='red')
plt.scatter(x2, y2, label=f'UserInput', color='green')
plt.title('Plot of two quadratic functions')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.show()

```
